```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
path1 = "data/SARSA/Portals/"
path2 = "data/Q_learning/Portals/"
titles = ['Close','Mid','Far']
for i in range(3):
    S_timeSteps_list,S_TotalReward_list,S_Episodes_list  = np.load(path1+f"{titles[i]}.npy")
    Q_timeSteps_list,Q_TotalReward_list,Q_Episodes_list  = np.load(path2+f"{titles[i]}.npy")
    #save into files
    f, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6), sharey=False)
    # plotting the points
    ax1.plot(S_Episodes_list, S_TotalReward_list,label="SARSA")
    ax2.plot(S_timeSteps_list,label="SARSA")
    
    ax1.plot(Q_Episodes_list, Q_TotalReward_list,label="Q learning")
    ax2.plot(Q_timeSteps_list,label="Q learning")
    
    print(f"Final optimal path found by SARSA is of lenght {S_timeSteps_list[-1]}")
    print(f"Final optimal path found by Q learning is of lenght {Q_timeSteps_list[-1]}")
    # naming the x axis
    ax1.set_xlabel('Episodes')
    ax2.set_xlabel('Episodes')
    # naming the y axis
    ax1.set_ylim(-1,1.2)
    ax1.set_ylabel('Total Reward')
    ax2.set_ylabel('Time steps to solve')

    # giving a title to my graph
    ax1.set_title(f'Reward for the "{titles[i]}" goal')
    ax2.set_title(f'Steps taken to solve for the "{titles[i]}" goal')
    
    f.tight_layout()
    figtitle = f"Results for Maze with Portals for the {titles[i]} Goal"
    f.suptitle(figtitle,fontsize=20,y=1.05)
    plt.legend()
    plt.savefig("figs/"+figtitle)
    plt.show()
```
